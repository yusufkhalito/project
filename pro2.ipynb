```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
```

```python
df = pd.read_csv('StudentsPerformance.csv')
df.head()
```

```python
fig = px.box(df, x ='gender', y ='math score')
fig.update_layout(
    xaxis_title='Gender',
    yaxis_title='Math score',
    title='Gender distribution vs Math score'
)
fig.show()
```

```python
fig = px.bar(df, x ='parental level of education', y = 'math score',color = 'gender', text='lunch')
fig.show()
```

```python
fig = px.bar(df, x ='lunch', y = 'math score',color = 'writing score', text='lunch')
fig.show()
```

```python
plt.figure(figsize=(12, 8))
ax = sns.swarmplot(x='race/ethnicity',y='writing score',data=df,palette='Blues', hue='gender')

```

```python
plt.figure(figsize=(12, 8))
ax = sns.violinplot(x='lunch',y='math score',data=df,palette='Blues', hue='gender')

```

```python
fig = px.box(df, x ='test preparation course', y ='reading score',color = 'gender')
fig.show()
```

```python
sns.boxplot(x='test preparation course', y='reading score',data=df)
plt.title('Reading Scores based on Test Preparation Course Completion')
plt.xlabel('Test Preparation Course Completion (0: Not Completed, 1: Completed)')
plt.ylabel('Reading Score')
plt.show()
```

```python
fig = px.bar(df, x ='race/ethnicity', y = 'reading score',color = 'gender', text='lunch')
fig.show()
```

```python
sns.boxplot(x='lunch', y='writing score', data=df)
plt.title('Boxplot of Writing Scores by Lunch Type')
plt.show()
```

```python
fig = px.histogram(df, x='parental level of education',text_auto=True, color='writing score')
fig.show()
```
